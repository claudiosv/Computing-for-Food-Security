Filter for Item Codes that indicate group totals

```python
item_totals = all_years[(all_years["Item Code"] >= 2902) & (all_years["Item Code"] < 3000)]
```

Check the possible elements related to food production and usage

```python
item_totals[item_totals["Unit"] == "1000 tonnes"].Element.unique()
```

```python
item_totals[(item_totals["Unit"] == "1000 tonnes") & (item_totals["Year"] == "2010")].groupby("Element").Value.agg("sum")
```

From the FOA sidebar: 

The total quantity of foodstuffs produced in a country (**Production**) added to the total quantity imported (**Import Quantity**) and adjusted to any change in stocks (**Stock Variation**) that may have occurred since the beginning of the reference period gives the supply available during that period. 

On the utilization side a distinction is made between the quantities exported (**Export Quantity**), fed to livestock (**Feed**), used for seed (**Seed**), put to manufacture for food use (**Processing**) and non-food uses (**Other uses (non-food)**), losses during storage and transportation (**Losses**), and food supplies available for human consumption.

More details on definitions: https://www.fao.org/3/x9892e/X9892e02.htm#TopOfPage

Production + imports - exports + changes in stocks (decrease or increase) = supply for domestic utilization.

```python
tonnes = item_totals[item_totals["Unit"] == "1000 tonnes"]

# get supply/demand totals for each area by year
supply_consumption = pd.pivot_table(tonnes, index=["Area", "Year"], columns=["Element"], values=["Value"], aggfunc="sum")
```

```python
supply_consumption.columns = supply_consumption.columns.droplevel()
```

```python
supply_consumption.reset_index().to_csv("consumption.csv", index=False)
supply_consumption
```
